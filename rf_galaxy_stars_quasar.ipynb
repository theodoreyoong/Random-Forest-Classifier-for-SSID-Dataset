_The Sloan Digital Sky Survey_ has created the most detailed three-dimensional maps of the Universe ever made, with deep multi-color images of one third of the sky, and spectra for more than three million astronomical objects.

Definitions:

Quasar- Also known as quasi-stellar object, it is an extremely luminous active galactic nucleus (AGN). The power radiated by quasars is enormous. The most powerful quasars have luminosities exceeding 1041 watts, thousands of times greater than an ordinary large galaxy such as the Milky Way.

Galaxy- A galaxy is a gravitationally bound system of stars, stellar remnants, interstellar gas, dust, and dark matter. Galaxies are categorized according to their visual morphology as elliptical, spiral, or irregular. Many galaxies are thought to have supermassive black holes at their active centers.

Star- A star is type of astronomical object consisting of a luminous spheroid of plasma held together by its own gravity. The nearest star to Earth is the Sun.

Brief Details of each observation:

ra (right ascension) and dec (declination) are the coordinates on the sky that correspond to longitude and latitude on Earth. RA measures east and west on the celestial sphere and is like longitude on the Earth. Dec measures north and south on the celestial sphere and is like latitude on the Earth.

objid = Object Identifier

ra = J2000 Right Ascension (r-band)

dec = J2000 Declination (r-band)

Knowing that the color of a star is related to the wavelength of light it gives off, lets you understand the meaning of the SDSS's color filters. Each filter is designed to let in light around a specific wavelength. The filters work by blocking out light at all wavelengths except those around the wavelength they are designed to see.

u (ultraviolet)= better of DeV/Exp magnitude fit

g (green) = better of DeV/Exp magnitude fit

r (red) = better of DeV/Exp magnitude fit

i (Near infrared) = better of DeV/Exp magnitude fit

z (Infrared) = better of DeV/Exp magnitude fit

run = Run Number

rereun = Rerun Number
camcol = Camera column
field = Field number
specobjid = Object Identifier
class = object class (galaxy, star or quasar object)
redshift = Final Redshift
plate = plate number
mjd = MJD of observation
fiberid = fiber ID

Next few plots shows the density distribution of each column.

```python
gal = df[df['class']=='GALAXY']
star = df[df['class']=='STAR']
quas = df[df['class']=='QSO']
```

```python
pal = sns.color_palette()

plt.hist(gal['ra'], bins=200, range=[0, 200], color='indigo', alpha = 0.7, density=True, label='Galaxy')
plt.hist(star['ra'], bins=200, range=[0, 200], color='#FF69B4', alpha = 0.7, density=True, label='Star')
plt.hist(quas['ra'], bins=200, range=[0, 200], color='cyan', alpha = 0.7, density=True, label='Quasar')
plt.title('Density of Occurences', fontsize=15)
plt.legend()
plt.xlabel('Right ascension', fontsize=15)
plt.ylabel('Density', fontsize=15)
```

Density plot is smooth. I want that. We use kernel density estimation.

In this method, a continuous curve (the kernel) is drawn at every individual data point and all of these curves are then added together to make a single smooth density estimation. The kernel most often used is a Gaussian (which produces a Gaussian bell curve at each data point).

```python
# Iterate through the three labels
sns.distplot(gal['ra'], hist = False, kde = True, color='indigo', kde_kws = {'shade': True, 'linewidth': 3}, label = 'Galaxy')
sns.distplot(star['ra'], hist = False, kde = True, color='#FF69B4', kde_kws = {'shade': True, 'linewidth': 3}, label = 'Star')
sns.distplot(quas['ra'], hist = False, kde = True, color='cyan', kde_kws = {'shade': True, 'linewidth': 3}, label = 'Quasar')
    
# Plot formatting
plt.legend()
plt.title('Density Plot')
plt.xlabel('ra')
plt.ylabel('Density')
```

We've got the features and labels dataframes.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=123, stratify=labels)
```

```python
y_test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
```

```python
from pprint import pprint
pprint(clf.get_params())
```

```python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 100, num = 10)]

hyperparameters = {'max_features':[None, 'auto', 'sqrt', 'log2'],
                   'max_depth':[None, 1, 5, 10, 15, 20],
                   'min_samples_leaf': [1, 2, 4],
                   'min_samples_split': [2, 5, 10],
                   'n_estimators': n_estimators,
                   'criterion': ['gini', 'entropy']}
```

```python
rf_random = RandomizedSearchCV(estimator = clf,
                               param_distributions = hyperparameters,
                               n_iter = 100, cv = 10, verbose=2, random_state=123, n_jobs = -1)

rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
hyperparameters = {'max_features':[None, 'auto', 'sqrt', 'log2'],
                   'max_depth':[14, 15, 16],
                   'min_samples_leaf': [1, 2, 3],
                   'min_samples_split': [4, 5, 6],
                   'n_estimators': [90, 100, 110],
                   'criterion': ['entropy']}
```

```python
from sklearn.model_selection import GridSearchCV

rf_grid = GridSearchCV(clf, hyperparameters, cv = 10, n_jobs = -1, verbose = 2)
rf_grid.fit(x_train, y_train)
```

```python
rf_grid.best_params_
```

```python
clf.set_params(criterion = 'entropy', max_features = None, max_depth = 14, min_samples_leaf = 2,
               min_samples_split = 5, n_estimators = 100)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accscore = accuracy_score(y_test, y_pred)
print(accscore)
```

```python
from sklearn.metrics import f1_score
f1score = f1_score(y_test, y_pred, average = None)
print(f1score)
```

It is important to ensure that the way you label your confusion matrix rows and columns corresponds exactly to the way sklearn has coded the classes. The true order of the labels can be revealed using the .classes_ attribute of the classifier. You can use the code below to prepare a confusion matrix data frame.

```python
sortedlabels = clf.classes_
print(sortedlabels)
```

```python
for i in range(len(sortedlabels)):
    print((sortedlabels[i],f1score[i]), end=" ")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test, y_pred, sortedlabels)
print(classification_report(y_test, y_pred))
print(cm)
```

```python
cm = pd.DataFrame(cm, index=sortedlabels, columns=sortedlabels)

sns.set(font_scale=1.2)#for label size
sns.heatmap(cm, linewidths=0.5, cmap=sns.light_palette((1, 0.2, 0.6),n_colors=10000), annot=True)# font size
plt.xlabel('Predicted')
plt.ylabel('True')
```
